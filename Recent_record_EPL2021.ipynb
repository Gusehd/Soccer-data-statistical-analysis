## 축구 데이터 - 4. 최근 5 경기 전적으로 보는 경기 결과 EPL 20/21

```python
import pandas as pd

df_win = pd.DataFrame([[0, 2, 7, 6, 4, 2],[1, 8, 18, 23, 11],[3, 31, 38, 16],[22, 22, 16],[11, 7],
                       [10]],
                   index=['WIN 0', 'WIN 1' , 'WIN 2' , 'WIN 3' ,'WIN 4','WIN 5'],
                   columns=['LOSE 0', 'LOSE 1','LOSE 2','LOSE 3','LOSE 4','LOSE 5',])
df_draw = pd.DataFrame([[0, 1, 5, 5, 4, 2],[0, 5, 15, 11, 6],[2, 14, 16, 13],[9, 16, 8],[8, 5],
                        [1]],
                   index=['WIN 0', 'WIN 1' , 'WIN 2' , 'WIN 3' ,'WIN 4','WIN 5'],
                   columns=['LOSE 0', 'LOSE 1','LOSE 2','LOSE 3','LOSE 4','LOSE 5',])
df_lose = pd.DataFrame([[1, 2, 5, 11, 17, 4],[0, 8, 25, 24, 15],[5, 21, 36, 26],[7, 19, 18],
                        [7, 4],[1]],
                   index=['WIN 0', 'WIN 1' , 'WIN 2' , 'WIN 3' ,'WIN 4','WIN 5'],
                   columns=['LOSE 0', 'LOSE 1','LOSE 2','LOSE 3','LOSE 4','LOSE 5',])
df_all = pd.DataFrame([[1, 5, 17, 22, 25, 8],[1, 21, 58, 58, 32],[10, 66, 90, 55],[38, 57, 42]
                       ,[26, 16],[12]],
                   index=['WIN 0', 'WIN 1' , 'WIN 2' , 'WIN 3' ,'WIN 4','WIN 5'],
                   columns=['LOSE 0', 'LOSE 1','LOSE 2','LOSE 3','LOSE 4','LOSE 5',])
```

```python
df_win
```

### Win - 이번 경기를 이겼을 때 지난 5경기의 전적 확인
* 최근전적에서 나올수 있는 경우는 WIN / LOSE/ DRAW 의 3가지 경우이다.
* WIN + LOSE + DRAW = 5 ( 최근 전적 수 ) 이므로 위처럼 WIN 과 LOSE 만을 이용해 최근 전적을 두었다.
* 이번 경기를 이겼을 때 최근 전적을 살펴본 결과 38경기로 WIN-2 DRAW-1 LOSE-2 의 경우가 가장 많았다.

---

```python
df_draw
```

### Draw - 이번 경기를 비겼을 때 지난 5경기의 전적 확인
* 최근전적에서 나올수 있는 경우는 WIN / LOSE/ DRAW 의 3가지 경우이다.
* WIN + LOSE + DRAW = 5 ( 최근 전적 수 ) 이므로 위처럼 WIN 과 LOSE 만을 이용해 최근 전적을 두었다.
* 이번 경기를 비겼을 때 최근 전적을 살펴본 결과 16경기로 WIN-2 DRAW-1 LOSE-2 / WIN-3 DRAW-1 LOSE-1 의 경우가 가장 많다.

---

```python
df_lose
```

### Lose - 이번 경기를 졌을 때 지난 5경기의 전적 확인
* 최근전적에서 나올수 있는 경우는 WIN / LOSE/ DRAW 의 3가지 경우이다.
* WIN + LOSE + DRAW = 5 ( 최근 전적 수 ) 이므로 위처럼 WIN 과 LOSE 만을 이용해 최근 전적을 두었다.
* 이번 경기를 졌을 때 최근 전적을 살펴본 결과 36경기로 WIN-2 DRAW-1 LOSE-2 의 경우가 가장 많다.

---

```python
df_all
```

### EPL 20/21 시즌에서 발생한 모든 지난 5경기의 전적 확인
* 모든 전적을 살펴본 결과 최근 5경기 전적이 WIN-2 DRAW-1 LOSE-2 인 경우가 가장 많았다.
* 연속으로 5번을 모두 비긴 경우는 EPL 20/21 시즌을 통틀어 한 경기 밖에 없었다.
* 이 4가지 데이터 프레임을 통해 어떤 최근 전적에서 어떤 결과가 가장 많이 등장하는 지 확인해보도록 하자.

---

```python
df_win_rate = round(df_win/df_all, 2)
df_win_rate
```

### 해당하는 최근 전적을 가지고 있을 때 이번 경기를 이긴 비율
* 가장 많은 경우인 WIN-2 DRAW-1 LOSE-2 는 0.42의 비율을 보여주고 있다.
* 이외의 관측치가 작았던 WIN-1 DRAW-4 LOSE-0 의 경우는 비율이 1로 나오고 있다.
* 최근 전적에 승리 수가 많을 수록 비율이 점점 늘어나는 추세를 보이기는 한다.

---

```python
df_draw_rate = round(df_draw/df_all, 2)
df_draw_rate
```

### 해당하는 최근 전적을 가지고 있을 때 이번 경기를 비긴 비율
* 가장 많은 경우인 WIN-2 DRAW-1 LOSE-2 는 0.18의 비율을 보여주고 있다.
* 가장 많은 경우인 WIN-3 DRAW-1 LOSE-1 는 0.28의 비율을 보여주고 있다.
* 이외의 관측치가 아예 없었던 경우들은 모두 0의 비율로 나타나게 되었다.
* 대부분 0.2 ~ 0.3 사이의 값들에 분포되어 있는 것으로 보아 큰 관련은 없어보인다.

---

```python
df_lose_rate = round(df_lose/df_all, 2)
df_lose_rate
```

### 해당하는 최근 전적을 가지고 있을 때 이번 경기를 진 비율
* 가장 많은 경우인 WIN-2 DRAW-1 LOSE-2 는 0.40 의 비율을 보여주고 있다.
* 이외의 관측치가 아예 없었던 경우들은 모두 0의 비율로 나타나게 되었다.
* 관측치가 적어 비율이 들쭉날쭉 하지만 LOSE가 많을 수록 비율이 높게 나오는 경향은 보인다.

---

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [15, 10]
sns.heatmap(df_all, annot=True)
plt.title('발생한 전체 경우의 히트맵', fontsize=20)
plt.show() 
```

### 발생한 전체 최근전적 빈도 히트맵
* 히트맵을 보면 가장 많이 발생한 구간은 WIN-2 DRAW-1 LOSE-2 구간이고 그 주변에 분포되어 있음을 볼 수 있다.
* EPL 20/21 시즌의 대부분의 최근 5 경기 전적은 WIN은 1 ~ 3 / LOSE도 1 ~ 3 구간 사이에 머무르는 것이다.

---

```python
plt.rcParams['figure.figsize'] = [15, 10]
sns.heatmap(df_win_rate, annot=True)
plt.title('특정 최근전적에 대해 이긴 경우의 히트맵', fontsize=20)
plt.show() 
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
sns.heatmap(df_draw_rate, annot=True)
plt.title('특정 최근전적에 대해 비긴 경우의 히트맵', fontsize=20)
plt.show() 
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
sns.heatmap(df_lose_rate, annot=True)
plt.title('특정 최근전적에 대해 진 경우의 히트맵', fontsize=20)
plt.show() 
```
