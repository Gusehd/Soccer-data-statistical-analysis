* 1 -> normal / draw -> 0 / exp -> -1 / same -> -2 로 두고 dataframe을 구성함
* 이를 통해 각 라운드별 normal이 발생한 비율과 2 ~ 38 라운드까지 진행하면서 normal_rate가 어떻게 변화하는지 확인
* 이후 각라운드가 진행되면서 normal_rate의 평균이 어떻게 변화하는지 확인

---

```python
import math

sum_array = []
move_array = [0.4]

for t in range(0,37):
    sum = 0
    for i in df.loc[t]:
        if i == 1:
            sum += 1
    sum_array.append(sum/20)
#print(sum_array)
df["Round_Rate"] = sum_array

for i in range(1,37):
    ans = round(((i*move_array[i-1]+sum_array[i])/(i+1)),3)
    move_array.append(ans)
#print(move_array)

df["Mean_Rate"] = move_array
df
```

* Round_rate 를 통해서는 각 라운드별 normal이 얼마나 발생하였는지 비율로 확인
* Mean_rate 를 통해서는 각 라운드가 진행됨에 따라 전체적인 경기수에 비해 normal이 얼마나 발생하였는지 비율로 확인
